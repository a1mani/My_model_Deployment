Description

    The data consists of 4,165 ABO3 perovskite-type oxides. Each observation is described by 13 feature columns and 1 class column which identifies it to be either a cubic, tetragonal, orthorhombic, and rhombohedral structure.

    v(A) - Valence of A

    v(B) - Valence of B 

    r(AXII)(Å) or rX - ionic radius of the anion

    Radius A (r(A)) - ionic radius of A cation

    Radius B (r(B)) - ionic radius of B cation

    Electronegativity of A (EN(A)) - Average electronegativity value of A cation

    Electronegativity of B (EN(B)) - Average electronegativity value of B cation

    l(A-O) - Bond length of A-O pair

    l(B-O) - Bond length of B-O pair

    ΔENR - Electronegativity difference with radius

    tG - Goldschmidt tolerance factor

    τ - New tolerance factor 

    Μ - Octahedral factor


    cubic
    tetragonal
    orthorhombic

    rhombohedral

### df2.drop(['radius_A', 'radius_B'], axis=1, inplace=True)

```python
df2
```

```python
df2['Lowest distortion']=df2['Lowest distortion'].map({'cubic':0,'tetragonal':1,'orthorhombic':2 ,'rhombohedral':3})
```

```python
X=df2.iloc[:,:13]
y=df2['Lowest distortion']
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate=[]
for i in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
error_rate
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11),error_rate,color= 'blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
```

```python
knn = KNeighborsClassifier(n_neighbors=5, weights='uniform')
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,f1_score,classification_report,confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
score=f1_score(y_test,y_pred,average='micro')
score*100
```

### Hyper Parameter Tunning

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier (n_estimators=5)
rf_clf.fit(X_train,y_train)
```

```python
y_predict=rf_clf.predict(X_test)
print(classification_report(y_test,y_predict))
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = [int(x) for x in np.linspace(start=0, stop=800, num=10)]
max_features = ['auto','sqrt']
max_depth = [int(x) for x in np.linspace(50, 200, num=11)]
max_depth.append(None)
min_samples_split = [2,5,10]
min_samples_leaf  = [1,2,4]
bootstrap = [True, False]

random_grid = {'n_estimators': n_estimators ,'max_features':max_features ,'max_depth':max_depth ,
               'min_samples_split':min_samples_split , 'min_samples_leaf':min_samples_leaf ,'bootstrap':bootstrap}

rf_clf1 = RandomForestClassifier(random_state=42)

rf_cv = RandomizedSearchCV(estimator=rf_clf1, scoring='f1', param_distributions=random_grid, n_iter=100, cv=3,
                           verbose=3,random_state=42, n_jobs=-1)

rf_cv.fit(X_train, y_train)

rf_best_params = rf_cv.best_params_
print(f"Best parameters: {rf_best_params}")
```
